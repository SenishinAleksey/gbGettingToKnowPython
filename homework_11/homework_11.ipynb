```python
from sympy import *
x = Symbol('x', real=True)
f = x ** 2 - 6 * abs(x) + 8
f
```

```python
from sympy.plotting import plot
plot(f)
```

```python
roots = solve(f)
roots
```

```python
differ = diff(f)
differ
```

```python
plot(differ)
```

```python
roots_differ = solve(differ, real_roots=True)
roots_differ = [-3, 0] + roots_differ
roots_differ
```

```python
res = []
for root in roots_differ:
    res.append((root, f.subs(x, root)))
res
```

```python
solve(differ > 0)
```

```python
solve(differ < 0)
```

```python
solve(f > 0)
```

```python
solve(f < 0)
```
